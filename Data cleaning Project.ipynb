# Data Cleaning

## IMPORTING REQUIRED PACKAGES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## LOADING DATA FROM DATA SOURCE

```python
df = pd.read_excel("Customer Call List.xlsx")
df
```

```python
# Droping Duplicates
df = df.drop_duplicates()
df
```

```python
# Removing columns of not use
df =df.drop(columns="Not_Useful_Column")
df
```

# Data cleaning on last_name column

```python
# In last name we have some ...,/ and _ which is of not use and we have to remove them 
df["Last_Name"] = df["Last_Name"].str.strip("123.../_")
df
```

### Data cleaning on Phone_number column

```python
# Here we are removing all - and / from from our phone number and standarising it with regex expression
df["Phone_Number"]=df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
```

```python
# To perfom further operation on phone number we  have convert phone numbers to string for slicing purpose
df["Phone_Number"]=df["Phone_Number"].apply(lambda x:str(x))
df
```

```python
# Now in above steps we got our phone numbers in required form now in between numbers we are providing some characters
df["Phone_Number"]=df["Phone_Number"].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
# Here we removing all nan values from phone numbers
df["Phone_Number"]=df["Phone_Number"].str.replace('Na--','')
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('nan--','')
df
```

### Data Cleaning on Address column

```python
# Here we are splitting address as address contains this all three street state and zipcode into one so splitting it into three column
df[["street","state","zip_code"]]=df["Address"].str.split(',',n=2,expand=True)
df
```

```python
# Now address column is not useful as we created new columns with the help of it so droping address column
df = df.drop(columns="Address")
df
```

# Data cleaning on columns Paying Customer and	Do_Not_Contact	

```python
# Formating Paying Customer	Do_Not_Contact	 beacuse they have diffrent form of data yes as y and no as y
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
# Here we are filling NA values with blank\
df=df.replace('N/a','')
df=df.fillna('')
df
```

```python
# Now in do_not_call_contact column where there is y it means it saying do not contact to them so we are removing them
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == "Y":
        df.drop(x,inplace=True)
df
```

```python
# Now in phone_number  column where there is no number it means we can not call them so removing them all 

for x in df.index:
    if df.loc[x,"Phone_Number"]=='':
        df.drop(x,inplace=True)
df
```

```python
# Now adjust this data frames index 
df.reset_index(drop=True)
```
